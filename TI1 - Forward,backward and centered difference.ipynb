```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.misc import derivative
plt.style.use('seaborn-whitegrid')
```

# Methods

## 2 point forward difference

### $ f'(x_i) \approx \dfrac{f(x_{i+1}) - f(x_i)}{h}  $

```python
def forward_difference(f,x,h,true_value):
    results = np.array((f(x+h) - f(x)) / h)
    errors = results - true_value
    ratio_errors = errors[0:len(errors)-1] / errors[1: len(errors)]
    
    return results,errors, ratio_errors
```

## 2 point backward difference

### $ f'(x_i) \approx \dfrac{f(x_{i}) - f(x_{i-1})}{h}  $

```python
def backward_difference(f,x,h,true_value):
    results = np.array((f(x) - f(x - h)) / h)
    errors = results - true_value
    ratio_errors = errors[0:len(errors)-1] / errors[1: len(errors)]
    
    return results,errors, ratio_errors
```

## 2 point centered difference

### $ f'(x_i) \approx \dfrac{f(x_{i+1}) - f(x_{i-1})}{2h}  $

```python
def centered_difference(f,x,h,true_value):
    results = np.array((f(x + h) - f(x - h)) / (2*h))
    errors = results - true_value
    ratio_errors = errors[0:len(errors)-1] / errors[1: len(errors)]
    
    return results,errors, ratio_errors
```

# Examples

$ f(x) = e^{sen(x)} ; x = 0 ; f'(0) = 1 $

```python
f = lambda x: np.exp(np.sin(x))
x = 0
h = 1/2 ** np.arange(1,11) 
true_value = 1
```

```python
h
```

```python
results_f,errors_f,ratio_errors_f=forward_difference(f,x,h,true_value)
```

```python
results_b,errors_b,ratio_errors_b=backward_difference(f,x,h,true_value)
```

```python
results_c,errors_c,ratio_errors_c=centered_difference(f,x,h,true_value)
```

```python
plt.figure(figsize=(14,12))
plt.title('Convergence of Difference Methods ',size='x-large')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.plot(results_f,marker='v',ms=13.0,label='Forward')
plt.plot(results_b,marker='^',ms=13.0,label='Backward')
plt.plot(results_c,marker='o',ms=13.0,label='Centered')
plt.legend(title='Method:',title_fontsize='xx-large',fontsize='x-large')
plt.show()
```
